```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.sparse import lil_matrix, csc_matrix
from scipy.sparse.linalg import spilu, LinearOperator, cg

def create_finite_difference_matrix(nx, ny, dx, dy, Q, source_region, T_left, alpha):
    n = nx * ny
    A = lil_matrix((n, n))
    b = np.zeros(n)
    
    def index(i, j):
        return i + j * nx
    
    y_start_idx = int(source_region[0] / dy)
    y_end_idx = int(source_region[1] / dy)
    
    for j in range(ny):
        for i in range(nx):
            idx = index(i, j)
            if i == 0:
                if y_start_idx <= j <= y_end_idx:
                    A[idx, idx] = 1
                    b[idx] = T_left
                else:
                    A[idx, idx] = 1
                    b[idx] = 0
            elif i == nx - 1:
                A[idx, idx] = 1
                b[idx] = 0
            elif j == 0:
                A[idx, idx] = -1
                A[idx, index(i, j + 1)] = 1
                b[idx] = 0
            elif j == ny - 1:
                A[idx, idx] = -1
                A[idx, index(i, j - 1)] = 1
                b[idx] = 0
            else:
                A[idx, index(i - 1, j)] = alpha / dx**2
                A[idx, index(i + 1, j)] = alpha / dx**2
                A[idx, index(i, j - 1)] = alpha / dy**2
                A[idx, index(i, j + 1)] = alpha / dy**2
                A[idx, idx] = -2 * alpha / dx**2 - 2 * alpha / dy**2
                b[idx] = -Q
    
    return A.tocsc(), b

def solve_heat_equation(nx, ny, dx, dy, T_left, Q, source_region, alpha):
    A, b = create_finite_difference_matrix(nx, ny, dx, dy, Q, source_region, T_left, alpha)
    
    # Create preconditioner using incomplete LU decomposition
    M2 = spilu(A)
    M = LinearOperator(A.shape, M2.solve)
    
    # Solve using preconditioned conjugate gradient method
    x0 = np.zeros_like(b)
    tol = 1e-8
    maxiter = 1000
    result, info = cg(A, b, x0=x0, tol=tol, maxiter=maxiter, M=M, callback=print_iteration)
    
    if info > 0:
        print(f"Convergence achieved after {info} iterations.")
    elif info == 0:
        print("Convergence achieved successfully.")
    else:
        print(f"CG failed to converge after {maxiter} iterations.")
    
    T = result.reshape((ny, nx))
    return T

def print_iteration(xk):
    global iter_count
    iter_count += 1
    print(f"Iteration {iter_count}")

# Problem parameters
nx, ny = 50, 50
Lx, Ly = 1.0, 1.0
dx, dy = Lx / (nx - 1), Ly / (ny - 1)
T_left = 100.0
Q = 0.0
alpha = 1.0
source_region = (0.4, 0.6)

# Global iteration counter
iter_count = 0

# Solve the heat equation
T = solve_heat_equation(nx, ny, dx, dy, T_left, Q, source_region, alpha)

# Plot the results
X, Y = np.meshgrid(np.linspace(0, Lx, nx), np.linspace(0, Ly, ny))
plt.contourf(X, Y, T, 20, cmap='hot')
plt.colorbar(label='Temperature')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Temperature distribution')
plt.show()
```

```python
import numpy as np
from scipy.sparse.linalg import cg, LinearOperator
from scipy.sparse import csr_matrix

# Callback class to track iterations
class IterationCounter:
    def __init__(self):
        self.niter = 0

    def __call__(self, xk):
        self.niter += 1
        print(f"Iteration {self.niter}")

# Example sparse matrix and vector
size = 10
A = csr_matrix(np.random.rand(size, size))
b = np.random.rand(size)

# Simple diagonal preconditioner
M = np.diag(1 / np.diag(A.toarray()))
preconditioner = LinearOperator((size, size), matvec=lambda x: M @ x)

# Iteration counter instance
counter = IterationCounter()

# Solve using Conjugate Gradient method with preconditioner and callback
x, info = cg(A, b, M=preconditioner, callback=counter)

print(f"Solution: {x}")
print(f"Convergence info: {info}")
```
