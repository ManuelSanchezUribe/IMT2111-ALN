```python
import numpy as np
import scipy.sparse as sps
import scipy.sparse.linalg as spsl

def CG(A, b, x0=None, j_max=1000, tol=1e-10):
    """
    Implementation of conjugate gradients.
    """

    m, n = A.shape
    if x0 is None:
        x0 = np.zeros((m, 1))
    j = 0
    rj = (b - A@x0)
    rjm = (b - A@x0)
    xj = x0
    errs = [np.linalg.norm(rj)]
    while (np.linalg.norm(rj) > tol) and (j < j_max):
        j += 1
        if j==1:
            pj = rj
        else:
            tau = ((rj.T.dot(rj))/(rjm.T.dot(rjm)))[0, 0]
            pj = rj + tau*pj
        mu = ((rj.T.dot(rj))/(pj.T.dot((A @ pj))))[0, 0]
        xj = xj + mu*pj
        rjm = rj.copy()
        rj = rj - mu*A@pj
        errs.append(np.linalg.norm(rj))
    return xj, errs

def PCG(A, M_inv, b, x0=None, j_max=1000, tol=1e-10):
    """
    Implementation of preconditioned conjugate gradients.
    """

    m, n = A.shape
    if x0 is None:
        x0 = np.zeros((m, 1))
    j = 0
    rj = (b - A@x0)
    rjm = (b - A@x0)
    xj = x0
    zj = M_inv(rj)
    zjm = zj
    errs = [np.linalg.norm(rj)]
    while (np.linalg.norm(rj) > tol) and (j < j_max):
        j += 1
        if j==1:
            pj = zj
        else:
            tau = ((rj.T.dot(zj))/(rjm.T.dot(zjm)))[0, 0]
            pj = zj + tau*pj
        mu = ((rj.T.dot(zj))/(pj.T.dot((A @ pj))))[0, 0]
        xj = xj - mu*pj
        rjm = rj.copy()
        rj = rj - mu*A@pj
        zjm = zj.copy()
        zj = M_inv(rj)
        errs.append(np.linalg.norm(rj))
    return xj, errs

def Am1(m):
    """
    Matriz a del enunciado
    """

    b2 = (1/9)*np.ones(m-1)
    c2 = (1/9)*np.ones(m-1)
    b1 = (5/18)*np.ones(m)
    c1 = (5/18)*np.ones(m)

    B = sps.diags([b2, b1, b2], offsets=[-1, 0, 1])
    C = sps.diags([c2, c1, c2], offsets=[-1, 0, 1])
    I = sps.eye(m)

    A = sps.kron(B, I) + sps.kron(I, C)

    return A

def Am2(m):
    """
    Matriz a del enunciado
    """

    b2 = (-1)*np.ones(m-1)
    c2 = (-1)*np.ones(m-1)
    b1 = (2)*np.ones(m)
    c1 = (2)*np.ones(m)

    B = sps.diags([b2, b1, b2], offsets=[-1, 0, 1])
    C = sps.diags([c2, c1, c2], offsets=[-1, 0, 1])
    I = sps.eye(m)

    A = sps.kron(B, I) + sps.kron(I, C)

    return A

def SSOR(A, w):
    L = sps.diags(A.diagonal()) + w*sps.tril(A, k=-1)
    D = sps.diags(A.diagonal())
    L = sps.csr_matrix(L)
    def M_inv(x):
        x = spsl.spsolve_triangular(L, x)
        x = D@x
        x = spsl.spsolve_triangular(L.transpose(), x, lower=False)
        return x
    return M_inv
```

```python
m = 80
tol = 1e-8
A1 = Am1(m)
A2 = Am2(m)
b = np.random.randn(m*m, 1)

M1_inv_SGS = SSOR(A1, w=1)
M2_inv_SGS = SSOR(A2, w=1)
x1cg, errs1cg = CG(A1, b, tol=tol)
x2cg, errs2cg = CG(A2, b, tol=tol)
x1pcg, errs1pcg = PCG(A1, M1_inv_SGS, b, tol=tol)
x2pcg, errs2pcg = PCG(A2, M2_inv_SGS, b, tol=tol)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 10))
iters1cg = list(range(len(errs1cg)))
iters2cg = list(range(len(errs2cg)))
iters1pcg = list(range(len(errs1pcg)))
iters2pcg = list(range(len(errs2pcg)))
ax.plot(iters1cg, errs1cg, "--o", label="A1, CG")
ax.plot(iters2cg, errs2cg, "--o", label="A2, CG")
ax.plot(iters1pcg, errs1pcg, "--o", label="A1, PCG")
ax.plot(iters2pcg, errs2pcg, "--o", label="A2, PCG")
ax.legend()
ax.set_yscale("log")
ax.grid()
plt.show()
```

```python
# Finite Element Spaces
V = VectorH1(mesh, order=2, dirichlet="wall|inlet|cyl", dirichletz="fluid")
Q = H1(mesh, order=1)
X = V*Q
(u,p),(v,q) = X.TnT()

stokes = InnerProduct(Grad(u), Grad(v))*dx + div(u)*q*dx + div(v)*p*dx
a = BilinearForm(stokes)
a.Assemble()
c = Preconditioner(a, "local")
c.Update()

gf = GridFunction(X)
gfu, gfp = gf.components

f = LinearForm(X)
f += q*0*dx
f.Assemble()

uin = CF ( (1.5*4*y*(0.41-y)/(0.41*0.41), 0,0) )
gfu.Set(uin, definedon=mesh.Boundaries("inlet"))

res = gf.vec.CreateVector()
res.data = f.vec-a.mat * gf.vec
```

```python
numpy_vertices = np.load('vertices.npy')
number_of_points = numpy_vertices.shape[0]
numpy_elements = np.load('elements.npy')
number_of_elements = numpy_elements.shape[0]
```

```python
my_vtk_dataset = vtkUnstructuredGrid()

points = vtkPoints()
for id in range(number_of_points):
    points.InsertPoint(id, numpy_vertices[id,:])
my_vtk_dataset.SetPoints(points)

elements_type = VTK_TETRA
my_vtk_dataset.Allocate(number_of_elements)
for id in range(number_of_elements):
    point_ids = numpy_elements[id,:]
    my_vtk_dataset.InsertNextCell(elements_type, 4, point_ids)

array = vtkDoubleArray()
array.SetNumberOfComponents(3)
array.SetNumberOfTuples(number_of_points)
array.SetName('Displacement')
for id in range(number_of_points):
    values = [xsol[id], xsol[id+number_of_points], xsol[id+2*number_of_points]]
    array.SetTuple(id, values)
my_vtk_dataset.GetPointData().AddArray(array)
```

```python
writer = vtkXMLUnstructuredGridWriter()
writer.SetFileName("tarea6_displacement.vtu")
writer.SetInputData(my_vtk_dataset)
writer.Write()
```
